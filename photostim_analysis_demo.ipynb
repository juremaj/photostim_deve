```python
print(f.shape)
f_responsive = f[228, :]
f_responsive_2 = f[233, :]
print(f_responsive.shape)
```

```python
plt.figure(figsize=(20, 2))
plt.plot(f_responsive)
plt.plot(f_responsive_2)
```

```python
stim_tstamps = np.arange(76, 4200, 94)
print(stim_tstamps.shape)

plt.figure(figsize=(20, 2))
plt.plot(f_responsive)
plt.vlines(stim_tstamps, ymin=200, ymax=600, color='r', linestyle='--')
```

```python
responses = np.zeros((len(stim_tstamps), 100))
responses_2 = np.zeros((len(stim_tstamps), 100))
for i, t in enumerate(stim_tstamps):
    responses[i, :] = f_responsive[t-20:t+80]
    responses_2[i, :] = f_responsive_2[t-20:t+80]
print(responses.shape)
```

```python

plt.plot(responses[:, :].T, color='gray', alpha=0.1)
plt.plot(np.mean(responses, axis=0))
plt.show()
plt.plot(responses_2[:, :].T, color='gray', alpha=0.1)
plt.plot(np.mean(responses_2, axis=0), c='C1')
plt.show()
```

```python
plt.imshow(responses)
plt.axvline(20, color='r', linestyle='--')
plt.show()

plt.imshow(responses_2)
plt.axvline(20, color='r', linestyle='--')
plt.show()
```

```python
max_responses_1 = np.max(responses, axis=1)
max_responses_2 = np.max(responses_2, axis=1)
```
