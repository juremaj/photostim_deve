# Quick and dirty code for plotting NIRB zstack for jm061 (data on desktop)

```python
import tifffile as tiff
import napari
import os
import numpy as np
```

```python
path_1100nm = '/Users/jure/Desktop/2025-11-02_z/1100nm/TSeries-10032023-1822-015'
path_830nm = '/Users/jure/Desktop/2025-11-02_z/830nm/TSeries-10032023-1822-016'
```

```python
# now for 1100nm first - find all the tiff files in the folder
all_tiff_1100nm = [f for f in os.listdir(path_1100nm) if f.endswith('.tif')]
all_tiff_1100nm.sort()  # sort to ensure correct order
fov_tseries_1100 = [tiff.imread(os.path.join(path_1100nm, f)) for f in all_tiff_1100nm]
fov_tseries_1100 = np.array(fov_tseries_1100)
```

```python
# now average across repetitions for each day
fov = np.mean(fov_tseries_1100, axis=0)  # shape should be (6, height, width)
print(fov.shape)  # should print (6, height, width)
```

```python
# now open in napari
viewer = napari.Viewer()
viewer.add_image(fov, name='1100nm zstack', scale=(3,1,1), colormap='red')  # adjust scale as needed
napari.run()
```
