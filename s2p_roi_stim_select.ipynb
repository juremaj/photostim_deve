```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data_path = 'data_proc/jm/jm048/2025-06-05_a/'
```

```python
s2p_path = os.path.join(data_path, 'suite2p', 'plane0')
```

```python
# load the ops

ops = np.load(os.path.join(s2p_path, 'ops.npy'), allow_pickle=True).item()
stat = np.load(os.path.join(s2p_path, 'stat.npy'), allow_pickle=True)
iscell = np.load(os.path.join(s2p_path, 'iscell.npy'), allow_pickle=True)
```

```python
# now filter stat based on the iscell manual curation
iscell_bool = iscell[:,0].astype(bool)
print(sum(iscell_bool))
```

```python
suite2p_idxs_iscell = np.arange(iscell.shape[0])[iscell_bool]
```

```python
stat_iscell = []
meds = []
for i, s in enumerate(stat):
    if iscell_bool[i]:
        stat_iscell.append(s)
        meds.append(s['med'])
```

```python
mn_image = ops['meanImg']
```

```python
# scatter the medians
meds = np.array(meds)
plt.figure(figsize=(5, 5), dpi=300)
plt.scatter(meds[:,1], meds[:,0], c='C0', s=1)
plt.imshow(mn_image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.title('Medians of all cells')
```

```python
# now randomly choose 30 cells
n_cells_choose = 30
seed = 123

np.random.seed(seed)
inds = np.random.choice(len(meds), n_cells_choose, replace=False)
print(inds)
```

```python
# now plot the cells
# scatter the medians
meds = np.array(meds)
plt.figure(figsize=(5, 5), dpi=300)
plt.scatter(meds[:,1], meds[:,0], c='C0', s=1)
for i in inds:
    plt.scatter(meds[i,1], meds[i,0], c='C1', s=1)
    # add text of index and x and y coordinates
    plt.text(meds[i,1], meds[i,0]+5, suite2p_idxs_iscell[i], fontsize=5, color='C1', ha='right', va='top')
plt.imshow(mn_image, cmap='gray', vmin=0, vmax=180)
plt.axis('off')
plt.title('Medians of all cells')
```

```python
# print the coordinates of the medians of the cells
print('Medians of the ROIs (in randomised order):')
# randomise the order of the indices
inds = np.random.permutation(inds)
for i in inds:
    # add leading zeros to the index
    i_str = str(i).zfill(3)
    print(f'ROI {i_str}: {meds[i]}')
```
